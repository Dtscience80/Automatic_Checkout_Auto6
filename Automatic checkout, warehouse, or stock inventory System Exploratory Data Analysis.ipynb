1. Data Cost (Cost_df)

```python
#Memuat data Cost dari google drive
cost_df = pd.read_csv ('/content/drive/MyDrive/AUTO 6/Costs AUTO6.csv' , sep = ";")
cost_df.head() #menampilkan 5 baris teratas data set
```

Identifikasi Bentuk Dataset

```python
cost_df.shape #menampilkan bentuk dataset (baris,kolom)
```

Display Data Set

```python
display (cost_df)
```

Identifikasi Nama Kolom Pada Dataset

```python
cost_df.columns #menampilkan nama kolom
```

Identifikasi Tipe Data Pada Tiap Kolom Dataset

```python
cost_df.dtypes #menampilkan tipe data tiap kolom
```

Identifikasi Missing Value

```python
print('Missing Values :',cost_df.isnull().values.any()) #menampilkan missing value (data yang na/null atau tidak terdeteksi/kosong pada data set)
print(cost_df.isnull().sum().sort_values(ascending=False)) #menampilkan missing value pada tiap kolom
```

Identifikasi Data Duplicate

```python
print('\Berikut jumalah data duplicate:')
cost_df[cost_df.duplicated(keep=False)] #menampilkan data yang duplicate - output : tidak ada data yang duplicate
```

Describe Dataset

```python
cost_df.describe() # deskripsi data cost
```

Visualisasi Data

Identifikasi Kolerasi Data Cost Dengan Heatmap

```python
sns.heatmap(data=cost_df.corr()) #menampilkan kolerasi antar data, warna muda kolerasi makin tinggi, sedangkan warna pekat kolerasi makin rendah
```

Kolerasi Menggunakan Line Plot

```python
cost_df.plot()
plt.tight_layout()
```
